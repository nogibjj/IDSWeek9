```python
import pandas as pd
```

```python

url = "https://raw.githubusercontent.com/reisanar/datasets/master/HollywoodMovies.csv"
data = pd.read_csv(url)
data.head()
```

```python
movies_after_2010 = data[data['Year'] > 2010]
movies_after_2010
```

```python
sorted_by_profitability = data.sort_values('Profitability', ascending=False)
sorted_by_profitability.head()
```

```python
average_rotten_tomatoes_score = data['RottenTomatoes'].mean()
average_rotten_tomatoes_score
```

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(data['Budget'], data['Profitability'], alpha=0.5)
plt.title('Scatter plot of Profitability vs. Budget')
plt.xlabel('Budget')
plt.ylabel('Profitability')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10,6))
plt.scatter(data['Year'], data['Profitability'], alpha=0.5)
plt.title('Scatter plot of Profitability vs. Year')
plt.xlabel('Year')
plt.ylabel('Profitability')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10,6))
plt.scatter(data['RottenTomatoes'], data['Profitability'], alpha=0.5)
plt.title('Scatter plot of Profitability vs. Rotten Tomatoes')
plt.xlabel('Rotten Tomatoes')
plt.ylabel('Profitability')
plt.grid(True)
plt.show()
```
